```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

#### 1. Use the API to get all available data for the _GDP per capita, PPP (constant 2017 international $)_ indicator. Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.

```python
endpoint = 'http://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?format=json'  ### added '?format=json' at end to convert to json from XML

params = {
    "per_page": "17290"
}
```

```python
response = requests.get(endpoint, params = params)

response    ##checking response to make sure its not 404
```

```python
res = response.json()
res
```

```python
records = res[1]

##gdp_df = pd.DataFrame(records)
```

```python
##gdp_df
```

```python
gdp_df = pd.json_normalize(records)
```

```python
gdp_df
```

```python
gdp_df = gdp_df.drop(columns=["unit", "obs_status", "decimal"], errors='ignore')
gdp_df
```

```python
gdp_df = gdp_df.rename(columns={
    "countryiso3code": "country_code",
    "value": "gdp_value"
})

gdp_df
```

#### 2. Now, use the API to get all available data for _Life expectancy at birth, total (years)_. This indicator has code "SP.DYN.LE00.IN". Again, convert the results to a DataFrame.

```python
endpoint = 'http://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN?format=json'

params = {
    "per_page": "18000"
}
```

```python
response = requests.get(endpoint, params = params)

response
```

```python
life_res = response.json()
life_res
```

```python
life_records = life_res[1]

###life_expectancy_df = pd.DataFrame(life_records)
life_expectancy_df = pd.json_normalize(life_records)
life_expectancy_df
```

```python
life_expectancy_df = life_expectancy_df.drop(columns=["unit", "obs_status", "decimal"], errors='ignore')

life_expectancy_df = life_expectancy_df.rename(columns={
    "countryiso3code": "country_code",
    "value": "life_expectancy_value"
})

life_expectancy_df
```

#### 3. Merge the two results DataFrames together. You may want to rename or drop columns prior to merging.

```python
merged_df = pd.merge(
    life_expectancy_df,
    gdp_df,
    on=['country_code', 'date', 'country.value', 'country.id'])   ### merged on multiple columns instead of just one
    ## how="inner"  ##reminder- automatically does inner


merged_df
```

#### 4. You can also get more information about the available countries (region, capital city, income level classification, etc.) by using the [Country API](https://datahelpdesk.worldbank.org/knowledgebase/articles/898590-country-api-queries). Use this API to pull in all available data. Merge this with your other datasets. Use this to now remove the rows that correspond to regions and not countries.

```python
endpoint = 'https://api.worldbank.org/v2/country?format=json'

params = {
    "per_page": "500"
}

response = requests.get(endpoint, params = params)

response
```

```python
country_res = response.json()

country_res
```

```python
all_countries_records = country_res[1]

##all_countries_df = pd.DataFrame(all_countries_records)

all_countries_df = pd.json_normalize(all_countries_records)

all_countries_df
```

```python
### in the following two inputs I renamed id to country_code in order to merge all dfs on country_code. instead, i could have done a left join on id and joined with country code which would have preserved the capitalCity column that was used later to fi;ter the results and get the 13K rows.
```

```python
all_countries_df= all_countries_df.rename(columns= {
        "id": "country_code"
})
```

```python
all_merged_df = pd.merge(
    merged_df,
    all_countries_df,
    on="country_code",
    how="inner"
)

all_merged_df
```

```python
only_countries = all_merged_df.loc[all_merged_df['capitalCity'] != '']
only_countries
```
